# General time functions
First set up al the general functions used in the analysis

```python
#Define general functions
# Convert times to minutes

def hms_to_s(s):
    t = 0
    for u in s.split(':'):
        t = 60 * t + int(u)
    return t
 
def hms_to_min(s):
  s=s.split(':')
  t = int(s[0])*60+int(s[1])*1 +int(s[2])/60
  return t
 
def ms_to_min(s):
  s=s.split(':')
  t = int(s[0])*1 +int(s[1])/60
  return t


def Clean_times(times):
    # Search for opening bracket in the time followed by
    # any characters repeated any number of times
    if re.search('\(.*', times):
        # Extract the position of beginning of pattern
        pos = re.search('\(.*', times).start()
        # return the cleaned time
        return times[:pos]
    else:
        # if clean up needed return the same time
        return times


def Split_times(times):
    split_re = re.compile('.*\((.*)\).*')
    # Search for opening bracket in the time followed by
    # any characters repeated any number of times
    if re.search('\(.*', times):
        # Split the times in group based on the pattern
        match = re.search(split_re, times)
         # return the cleaned time between parentheses
        return match.group(1)
    else:
        # if clean up needed return the same time
        return times
    
def convert(seconds):
    seconds = seconds % (24 * 3600)
    hour = seconds // 3600
    seconds %= 3600
    minutes = seconds // 60
    seconds %= 60
    return "%d:%02d:%02d" % (hour, minutes, seconds)

def hms_to_time(s):
    t = 0
    for u in s.split(':'):
        t = 60 * t + int(u)
    return convert(t)

def seconds_to_time(time_to_convert):
     return datetime.strptime(convert(time_to_convert), '%H:%M:%S').time()
```

# splittimes
Splittimes are the value in parenthesis. The distance times are the value before that. These two values need to be separated.
The Rotterdam Marathon uses 5Km splits, so in total 8 splits with distance of 5km and 1 split for the last 2,195km.

After splittting the times, they are converted to a time in seconds to allow calculations and comparisons.

```python
#Split the distance times from the split times
splits = ("5KM","10KM","15KM","20KM","25KM","30KM","35KM","40KM")

for i in splits:
    runners_df[['distance_'+i+'_time','split_'+i+'_time']] = runners_df[i].apply(lambda x: pd.Series(str(x).split(" ")))
    runners_df.loc[:,'distance_'+i+'_time'] = runners_df.loc[:,'distance_'+i+'_time'].astype(str).apply(Clean_times)
    runners_df.loc[:,'split_'+i+'_time'] = runners_df.loc[:,'split_'+i+'_time'].astype(str).apply(Split_times)
    runners_df[['distance_'+i+'_time','split_'+i+'_time']] = runners_df[['distance_'+i+'_time','split_'+i+'_time']].replace({'nan':np.nan, 0:np.nan, "":np.nan}).fillna("00:00")
```

```python
#Convert all the time strings to a time in seconds
cols=[]
for i in splits:
    t1 = 'split_'+i+'_time'
    t2 = 'distance_'+i+'_time'
    cols.append(t1)
    cols.append(t2)

for i in cols:
    runners_df[i]= runners_df[i].apply(hms_to_s)
runners_df["Guntime_s"] = runners_df["Guntime"].apply(hms_to_s)
runners_df["Chiptime_s"] = runners_df["Chiptime"].apply(hms_to_s)
runners_df["Halftime_s"] = runners_df["HALF"].apply(hms_to_s)

#Calculate last 2195m split
runners_df["split_42KM_time"] = runners_df["Chiptime"].apply(hms_to_s) - runners_df["distance_40KM_time"]
runners_df["distance_42KM_time"] = runners_df["Chiptime"].apply(hms_to_s)
runners_df.head(10)
```

For 2009-2017, using the ISO3166 standard, the 3-letter FIFA codes are converted to 3-letter ISO codes.
https://simple.wikipedia.org/wiki/Comparison_of_IOC,_FIFA,_and_ISO_3166_country_codes

```python
# get the data for the different categories
boxplot = pd.DataFrame()

boxplot['group'] = df_split['distance'].astype(str) + "KM"
boxplot['score'] = df_split.loc[df_split["split_time"]>0,["split_time"]]
cats = boxplot['group'].unique().tolist()

 

# find the quartiles and IQR for each category

groups = boxplot.groupby('group', sort=False)

q1 = groups.quantile(q=0.25)
q2 = groups.quantile(q=0.5)
q3 = groups.quantile(q=0.75)
iqr = q3 - q1
upper = q3 + 1.5*iqr
lower = q1 - 1.5*iqr
qmean = groups.mean()


 

# find the outliers for each category

def outliers(group):
    cat = group.name
    return group[(group.score > upper.loc[cat]['score']) | (group.score < lower.loc[cat]['score'])]['score']
out = groups.apply(outliers).dropna()


# prepare outlier data for plotting, we need coordinates for every outlier.

if not out.empty:
    outx = []
    outy = []
    for keys in out.index:
        outx.append(keys[0])
        outy.append(out.loc[keys[0]].loc[keys[1]])


# if no outliers, shrink lengths of stems to be no longer than the minimums or maximums

qmin = groups.quantile(q=0.00)
qmax = groups.quantile(q=1.00)
upper.score = [min([x,y]) for (x,y) in zip(list(qmax.loc[:,'score']),upper.score)]
lower.score = [max([x,y]) for (x,y) in zip(list(qmin.loc[:,'score']),lower.score)]
qmean.score = [max([x,y]) for (x,y) in zip(list(qmean.loc[:,'score']),qmean.score)]

p = figure(tools="", 
           background_fill_color="white", 
           x_range=cats, 
           toolbar_location=None, 
           title="Whiskerplot of the splittimes",
          )


# stems
p.segment(cats, upper.score, cats, q3.score, line_color="#ff7f0e")
p.segment(cats, lower.score, cats, q1.score, line_color="#ff7f0e")

# boxes
p.vbar(cats, 0.7, q2.score, q3.score, fill_color="#1f77b4", line_color="#ff7f0e")
p.vbar(cats, 0.7, q1.score, q2.score, fill_color="#1f77b4", line_color="#ff7f0e")


# whiskers (almost-0 height rects simpler than segments)
p.rect(cats, lower.score, 0.2, 0.01, line_color="#ff7f0e")
p.rect(cats, upper.score, 0.2, 0.01, line_color="#ff7f0e")

# outliers
#if not out.empty:
#    p.circle(outx, outy, size=6, color="#F38630", fill_alpha=0.6)
 

p.yaxis.formatter = NumeralTickFormatter(format='00:00:00')
#output_file("boxplot.html", title="boxplot.py example")
 
show(p)
```
